```python
#User Logic Check

username = "admin"
password = "1234"
input_user = input("Enter Username: ")
input_pass = input("Enter Password: ")

if input_user == username and input_pass == password:
    print("Login Successful!")
else:
    print("Invalid Credentials")
```

```python
#Pass/Fail Analyser

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Passed: ", pass_count)
print("Failed: ", fail_count)
```

```python
#Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
cleaned_list = []
for name in names:
    start = 0
    end = 0
    for i in range(len(name)):
        if name[i] != ' ':
            start = i
            break
    for i in range(len(name) - 1, -1, -1):
        if name[i] != ' ':
            end = i + 1
            break
    trimmed = name[start:end]
    lower_name = ""
    for char in trimmed:
        if 'A' <= char <= 'Z':
            lower_name = lower_name + chr(ord(char) + 32)
        else:
            lower_name = lower_name + char
    cleaned_list.append(lower_name)
print(cleaned_list)
```

```python
#Message Length Analyser

messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    count = 0
    for char in msg:
        count = count + 1
    print("Length:", count)
    if count > 10:
        print("Message too long")
```

```python
#Error Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
    if log == "ERROR":
        count = count + 1
print("Total Errors:", count)
```
