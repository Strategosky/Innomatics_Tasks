```python
#1. Unique Words
s=input("Enter sentence: ")
w=s.split()
u=set(w)
print(f"Unique words count: {len(u)}")
print(f"Unique words: {u}")
```

```python
#2. Highest Salary
e={"Ravi":75000,"Anita":68000,"Kiran":72000}
m=max(e,key=e.get)
print(f"Highest Salary: {m} - {e[m]}")
```

```python
#3. Max and Min
n=[45,22,89,10,66]
print(f"List: {n}")
print(f"Max: {max(n)}")
print(f"Min: {min(n)}")
```

```python
#4. Count Above Threshold
p=[450,1200,899,1500,300]
c=0
for x in p:
 if x>1000:
  c+=1
print(f"Products above 1000: {c}")
```

```python
#5. Attendance Percentage
a=["P","P","A","P","P"]
p=a.count("P")
print(f"Attendance Percentage: {p/len(a)*100}")
```

```python
#6. Unique Phone Numbers
ph=[9876543210,9123456789,9876543210]
print(f"Unique phone numbers: {set(ph)}")
```

```python
#7. Character Frequency
s="python"
d={}
for ch in s:
 d[ch]=d.get(ch,0)+1
print(d)
```
