```python
#1. Bonus Eligibility
employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}

max_score = max(employees.values())
top = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top)} (Score: {max_score})")
```

```python
#2. Search Keyword Analysis
query = "Buy mobile phone buy phone online"

counts = {}
for word in query.lower().split():
    counts[word] = counts.get(word, 0) + 1

result = {k:v for k,v in counts.items() if v > 1}
print(result)
```

```python
#3. Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = [ (hour, value) for hour, value in enumerate(sensor_readings) if value % 2 == 0 ]

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
#4. Email Domain Analysis
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]

counts = {}
total = len(emails)
for email in emails:
    domain = email.split('@')[1]
    counts[domain] = counts.get(domain, 0) + 1

for domain, count in counts.items():
    print(f"{domain}: {int(count * 100 / total)}%")
```

```python
#5. Sales Spike Detection
s = [1200, 1500, 900, 2200, 1400, 3000]
a = sum(s) / len(s)
for i, v in enumerate(s):
    if v > a * 1.3:  # more than 30% above average
        print(f"Day {i + 1}: {v}")
```

```python
#6. Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = {}
for user in user_ids:
    counts[user] = counts.get(user, 0) + 1

for user, count in counts.items():
    if count > 1:
        print(f"{user} → {count} times")
```
