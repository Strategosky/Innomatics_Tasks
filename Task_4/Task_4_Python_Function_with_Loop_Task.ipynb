```python
#1. Mobile Recharge Validation
def vr(a,p):return a>=50 and p in[199,299,399,599]
while 1:
    a=int(input("Amt: "))
    p=int(input("Plan: "))
    if vr(a,p):print("Valid");break
    print("Retry") # retry on invalid
```

```python
#2.  Inventory Reorder Alert
def ia(d):
    for k,v in d.items():print(f"{k}: {'Reorder Alert'if v<15 else'Stock OK'}")
ia({"A":10,"B":20})
```

```python
#3. Student Result Processing
def rp(m):
    t=c=0
    for x in m:t+=x;c+=1 # sum & count
    print("Pass"if t/c>=50 else"Fail")
    return t/c
rp([60,70,80])
```

```python
#4. Cab Fare Estimator
def fc(k,p):f=50+12*k;return f*1.25 if p else f
while 1:
    k=int(input("KM: "))
    p=input("Peak(y/n): ")=="y"
    print(f"Fare: ₹{fc(k,p)}")
    if input("Again? (y/n): ")=="n":break # retry loop
```

```python
#5. Attendance Eligibility
def ec(a):
    c=p=0
    for x in a:c+=1;p+=x=="P" # count total & present
    return"Eligible"if p/c*100>=75 else"Not Eligible"
print(ec(["P","P","A","P"]))
```

```python
#6. Password Strength
def ps(s):
    if len(s)<8:return"Weak"
    d=sp=0
    for c in s:
        if c.isdigit():d=1
        if c in"@#$":sp=1 # check digit & special
    return"Strong"if d and sp else"Weak"
print(ps("Innomatics%123"))
```
