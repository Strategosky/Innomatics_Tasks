```python
#1. Social Media Post Engagement Analyzer
def analyze_engagement(likes):
    total = sum(likes)               # loop to sum likes
    status = "Viral Post" if total >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total}\nPost Status: {status}")

# Example
analyze_engagement([200, 300, 550])
```

```python
#2.Healthcare Medicine Stock Alert System
def check_stock(qty):
    status = "Low Stock Alert" if qty < 10 else "Stock Sufficient"
    print(f"Medicine Stock: {qty}\nStatus: {status}")

# Example
check_stock(6)
```

```python
#3. Agriculture – Rainfall Adequacy Checker
def check_rainfall(data, required=50):
    total = sum(data)
    avg = total / len(data)
    st = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    print(f"Average Rainfall: {avg}\nRainfall Status: {st}")

# Example
check_rainfall([80, 70, 65, 75])
```

```python
#4. Social Media – Duplicate Account Detection
def detect_duplicates(names):
    seen, dup = set(), set()
    for n in names:
        dup.add(n) if n in seen else seen.add(n)
    print(f"Duplicate Accounts Found: {'Yes' if dup else 'No'}")

# Example
detect_duplicates(["alice", "bob", "alice", "charlie"])
```

```python
#5. Appointment Eligibility Checker
def eligible(age):
    print(f"Patient Age: {age}")
    print("Eligibility Status:", "Eligible" if age >= 18 else "Not Eligible")

#Example
eligible(21)
```

```python
#6. Premium Crop Price Filter
def filter_premium(prices, threshold=2000):
    premium = [p for p in prices if p > threshold]  # filter prices above threshold
    print(f"Premium Crops: {premium}")

#Example
filter_premium([1500, 2500, 1800, 3200])
```

```python
#7. Application Health Checker
def check_health(errors):
    if errors == 0: status = "Healthy"
    elif errors <= 5: status = "Minor Issues"
    else: status = "Critical Issues"
    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

#Example
check_health(7)
```

```python
#8. Daily Transaction Limit Checker
def check_transaction(amt, limit=50000):
    status = "Approved" if amt <= limit else "Rejected"
    print(f"Transaction Amount: {amt}")
    print(f"Transaction Status: {status}")

#Example
check_transaction(60000)
```

```python
#9. Student Attendance Eligibility
def check_attendance(present, total):
    pct = (present / total) * 100
    status = "Eligible" if pct >= 75 else "Not Eligible"
    print(f"Attendance Percentage: {pct}")
    print(f"Exam Eligibility: {status}")

#Example
check_attendance(80, 100)
```

```python
#10. Smart Electricity Bill Analyzer
def bill_calc(units):
    # Slab-based calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100 * 3 + (units - 100) * 5
    else:
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    print(f"Bill: ₹{bill}")
    print(f"Usage Level: {usage}")

#Example
bill_calc(250)
```
