```python
#1. Smart Parking Lot Management System
def smart_parking(cap, logs):
    c = 0
    for x in logs:
        if x == "IN":
            c += 1
        elif x == "OUT":
            c -= 1
        if c > cap:
            print("Alert: Capacity Exceeded")
    print("Currently Parked Vehicles:", c)
    print("Parking Status:", "Full" if c >= cap else "Available")


# example
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# 2. Online Food Delivery Time Estimator
def delivery_eta(d, traffic, weather):
    t = d * 5  # base: 5 min/km
    if traffic == "High":
        t += 15
    elif traffic == "Medium":
        t += 7
    if weather == "Rainy":
        t += 10
    elif weather == "Stormy":
        t += 20
    print(f"Estimated Delivery Time: {t} minutes")


# example
delivery_eta(8, "High", "Rainy")
```

```python
#3. Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total, booked):
    b = len(booked)
    occ = (b / total) * 100
    print(f"Occupancy: {int(occ)}%")
    if occ == 100:
        st = "Housefull"
    elif occ >= 75:
        st = "Almost Full"
    else:
        st = "Available"
    print("Show Status:", st)


# example
seat_occupancy(200, [1] * 150)
```

```python
#4. Cloud Server Load Classification System
def server_load(readings):
    s = 0
    for r in readings:
        s += r
    avg = s / len(readings)
    if avg < 50:
        st = "Normal"
    elif avg < 80:
        st = "Warning"
    else:
        st = "Critical"
    print(f"Average CPU Load: {int(avg)}%")
    print("Server Status:", st)


# example
server_load([45, 60, 70, 85, 90])
```

```python
#5. Smart Classroom Resource Usage Monitor
def resource_usage(u):
    over = []
    for k, v in u.items():
        if v > 8:
            over.append(k)
    print("Overused Resources:", ", ".join(over) if over else "None")
    print("Energy Alert:", "Yes" if over else "No")


# example
resource_usage({"Projector": 6, "AC": 9, "Lights": 4})
```

```python
#6. Online Event Registration Capacity Controller
def event_reg(cap, reg):
    if reg <= cap:
        conf = reg
        wait = 0
        st = "Open"
    else:
        conf = cap
        wait = reg - cap
        st = "Closed"
    print("Confirmed Registrations:", conf)
    print("Waitlisted Users:", wait)
    print("Registration Status:", st)


# example
event_reg(100, 105)
```
