```python
#1. Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display(self):
        print(f"{self.name} is {self.status}")

n = input("Light Name: ")
a = input("Action: ")
light = SmartLight(n)
if a == "ON":
    light.turn_on()
else:
    light.turn_off()
light.display()
```

```python
#2. Employee ID Card System
class Employee:
    def __init__(self, name, eid, dept):
        self.name = name
        self.eid = eid
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.eid}")
        print(f"Department: {self.dept}")

n = input("Employee Name: ")
i = input("Employee ID: ")
d = input("Department: ")
emp = Employee(n, i, d)
emp.display()
```

```python
#3. Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

n = input("Contact Name: ")
p = input("Phone Number: ")
c = Contact(n, p)
c.display()
```

```python
#4. Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

n = input("Product Name: ")
p = int(input("Price: "))
prod = Product(n, p)
prod.display()
```

```python
#5. Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

n = input("Movie Name: ")
r = float(input("Rating: "))
m = Movie(n, r)
m.display()
```

```python
#6. Delivery Address Manager
class Delivery:
    def __init__(self, name, addr):
        self.name = name
        self.addr = addr

    def display(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.addr}")

n = input("Customer Name: ")
a = input("Address: ")
d = Delivery(n, a)
d.display()
```
